```python
import pandas as pd
df_original=pd.read_csv("zomato_dataset.csv")
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df['delivery_time_min']=pd.to_numeric(df['delivery_time_min'],errors='coerce')
```

```python
df.head(20)
```

```python
df['delivery_time_min']=df['delivery_time_min'].fillna(60.0)
```

```python
df['order_value'].fillna(df['order_value'].median(),inplace=
                        True)
```

```python
df['discount_percent'].fillna(0,inplace=True)
```

```python
mapping = {'one':1,'two':2,'three':3,'four':4,'five':5}
df['rating'] = df['rating'].str.lower().str.strip()
df['rating'] = df['rating'].replace(mapping)
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df['rating'].fillna(0, inplace=True)
```

```python
df['cancellation_reason']=df['cancellation_reason'].fillna('No cancellation')
```

```python
df['refund_amount']=pd.to_numeric(df['refund_amount'],errors='coerce')
df['refund_amount'].fillna(0,inplace=True)
```

```python
df_clean=df.copy()
```

```python
df_clean.head(20)
```

```python
df['order_datetime']=pd.to_datetime(df['order_datetime'],errors='coerce')
```

```python
text_cols = ['restaurant_name','cuisine','payment_method','delivery_city','delivery_partner','food_items']

for col in text_cols:
    df[col] = df[col].astype(str).str.strip().str.title()
```

```python
df.loc[df['rating'] > 5, 'rating'] = 5
df.loc[df['rating'] < 0, 'rating'] = 0

df.loc[df['delivery_time_min'] < 0, 'delivery_time_min'] = 0
df.loc[df['delivery_time_min'] > 240, 'delivery_time_min'] = 240
```

```python
df['order_date'] = df['order_datetime'].dt.date
df['order_hour'] = df['order_datetime'].dt.hour
df['order_day'] = df['order_datetime'].dt.day_name()
```

```python
df['is_rated'] = df['rating'] > 0
```

```python
df.drop_duplicates(inplace=True)
```

```python
df = df.drop(columns=['user_lat', 'user_long', 'restaurant_lat', 'restaurant_long'])
```
